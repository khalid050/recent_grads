```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
grads = pd.read_csv('recent-grads.csv')
print(grads.head())
print(grads.tail())
print(grads.iloc[0])

```

```python
print(grads.describe())
```

```python
grads = grads.dropna()
grads.head()
```

```python
grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
grads.plot(x='Sample_size' , y='Unemployment_rate' , kind='scatter')
```

```python
grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
grads.plot(x='ShareWomen' , y='Unemployment_rate', kind='scatter')
```

```python
grads.plot(x='Men', y='Median', kind='scatter')
```

```python
grads.plot(x='Women', y='Median', kind='scatter')
```

```python
columns = ['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig = plt.figure(figsize=(8,16))
for i in range(1,6):
    ax = fig.add_subplot(5,1,i)
    ax = grads[columns[i]].plot(kind='hist')
    ax.set_xlabel(columns[i])
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(grads[['Sample_size' , 'Median']], figsize=(7,7))
```

```python
scatter_matrix(grads[['Sample_size','Median', 'Unemployment_rate']], figsize=(15,15))
```
